```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
dset=pd.read_csv('dataset1.csv')
dset
```

```python
range_fm=150
range_to=180
dset['Height']= np.random.randint(range_fm, range_to+1, len(dset))
print(dset['Height'])
```

```python
dset.head()
```

```python
emp_percentage=dset['Team'].value_counts()
emp_percentage
```

```python
percent_emp= pd.DataFrame(emp_percentage)
percent_emp
```

```python
percent_emp.rename(columns={'count':"No of employees"},inplace= True)
```

```python
percent_emp['Percentage of Employee']= percent_emp['count']/len(dset) * 100
```

## 2.Segregate the employees w.r.t different positions.

```python
dset['Position'].value_counts()
```

```python
segregated= dset.groupby('Position')
for Position,group in segregated:
    print(f"Employees in {Position} Position:")
    print(group)
    print("\n")
```

## 3.Find from which age group most of the employees belong to.

```python
most_age=dset['Age'].value_counts().idxmax()
print("Most of the Employees belongs to, age group:---",most_age)
```

## 4.Find out under which team and position, spending in terms of salary is high.

```python
high_salary= dset['Salary'].sort_values(ascending=False)
high_salary_max=high_salary.max()
result= dset.loc[dset['Salary']== high_salary_max,['Team','Position','Salary'] ]
result
```

## 5.Find if there is any correlation between age and salary , represent it visually.

```python
sns.set_style('darkgrid')
plt.scatter(dset['Salary'],dset['Age'])
plt.xlabel('Salary')
plt.ylabel('Age')
plt.title('Correlation Between Age and Salary')
plt.show()
```
